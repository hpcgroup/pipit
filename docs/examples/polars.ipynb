```python
def do_sum(df):
    stack = []
    matching_event = pd.Series(index=df.index, dtype="Int16")

    for i, event_type in df["Event Type"].items():
        if event_type == "Enter":
            stack.append(i)
        elif event_type == "Leave" and stack:
            m = stack.pop()
            matching_event.at[m] = i
            matching_event.at[i] = m

    df["_matching_event"] = matching_event
    return df

df["_matching_event"] = pd.Series(dtype="Int16")
df = df.groupby(['Process', 'Thread']).apply(do_sum).reset_index(drop=True)
```
